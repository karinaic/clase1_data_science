### Step 1. Import the necessary libraries

```python

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 2. Import the dataset of `chipotle` and assign it to a variable called chipo.

```python
address = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

chipo = pd.read_csv(address, sep = "\t")
chipo
```

### Step 3. See the first 10 entries

```python
chipo.head(10)
```

### Step 4. Create a histogram of the top 5 items bought

```python
#usamos el metodo de pandas groupby,aplicandola a la columna "iten name", con el metodo sum, para que nos devuelva la suma de cada uno de los valores de mayor a menor.
top_5 = chipo.groupby('item_name').quantity.sum().sort_values(ascending = False)[0:5]
top_5
```

```python
top_5.plot(kind='bar')

plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most ordered Chipotle\'s Items')
```

### Step 5. Create a scatterplot with the number of items orderered per order price

Make sure you get the same labels and title
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
orders = chipo.groupby('order_id').sum()
plt.scatter(x = orders.item_price, y = orders.quantity, s = 50, c = 'green')
plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price')
plt.ylim(0)
```
