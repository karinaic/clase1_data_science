### ReLu

```python
def ReLU(x):
    return (abs(x) + x) / 2
```

```python
ReLU(32626) # si da un valor positivo, siempre será ese número
```

```python
#ReLU(-32626) #si el valor es negativo, siempre sera "0"
```

### Sigmoid

```python
import math

# valor entre 0 y 1

def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
sigmoid(-590)
```

### Hyperbolic tangent

```python
import tensorflow as tf 
tf.keras.activations.tanh(2.) #lo que da es un valor entre 1 y -1
```
