
- [Cows_and_Bulls]

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python

x = "1234"
cow = 0
bulls = 0

while cow < 4:
    numero = input("Ingrese un número de 4 dígitos")
    cow = 0
    bulls = 0

    for i,value in enumerate(x):
        if value in numero and numero[i] == x[i]:
            cow += 1

        elif value in numero:
            bulls += 1

    print(" you got cow", cow , "you got bulls", bulls)

    
```
